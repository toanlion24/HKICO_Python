**Bài 9. Tính tổng 2**

Tính S(n) = 1^2 + 2^2 + 3^2 + ... + n^2 với n là số nguyên không âm (0 ≤ n ≤ 10^5).

Cách làm: sử dụng công thức tổng bình phương đã biết:  S(n) = n(n+1)(2n+1) / 6.
- Thời gian: O(1) (một vài phép toán số học).
- Ghi chú: với Python không lo tràn số vì int có độ rộng tùy ý lớn.

Ví dụ: Input: 100000 → Output: 333338333350000

```python

def sum_squares(n):
    return n * (n + 1) * (2 * n + 1) // 6

# Ví dụ kiểm thử nhanh
print('sum_squares(0) =', sum_squares(0))
print('sum_squares(1) =', sum_squares(1))
print('sum_squares(10) =', sum_squares(10))  # 385
print('sum_squares(100000) =', sum_squares(100000))  # 333338333350000

# Nếu dùng trong bài thi/đề:
# n = int(input())
# print(sum_squares(n))
```

```python
def tinh_tong_phan_so_sn():
    
    # Vòng lặp để yêu cầu nhập lại nếu có lỗi
    while True:
        # Bước 1: Nhập dữ liệu từ người dùng
        input_str = input("Mời bạn nhập số nguyên dương n (1 <= n <= 10^9): ")
        
        try:
            # Chuyển đổi chuỗi nhập vào thành số nguyên
            n = int(input_str)
            
            # Kiểm tra ràng buộc bài toán
            if n < 1:
                print("Lỗi: Số n phải là số nguyên DƯƠNG (>= 1).")
            elif n > 10**9:
                print("Lỗi: Số n quá lớn. Bài toán giới hạn n <= 10^9.")
            else:
                # Nếu nhập hợp lệ, thoát khỏi vòng lặp nhập
                break 
                
        except ValueError:
            # Xử lý nếu người dùng nhập chữ hoặc ký tự không phải số
            print("Lỗi: Đầu vào không hợp lệ. Vui lòng CHỈ nhập một số nguyên.")
            
    # Bước 2: Áp dụng công thức rút gọn O(1)
    # S_n = n / (n + 1)
    
    # Thực hiện phép chia số thực
    sn_value = n / (n + 1)
    
    # Bước 3: Định dạng kết quả lấy độ chính xác 2 chữ số thập phân
    # Sử dụng f-string formatting: :.2f
    result_output = f"{sn_value:.3f}"
    
    # Bước 4: In ra kết quả
    print("-" * 45)
    print(f"Giá trị n đã nhập: {n}")
    print(f"Tổng S_n = 1/(1*2) + ... + 1/({n}*({n}+1)) là:")
    print(f"KẾT QUẢ (làm tròn 2 chữ số): {result_output}")

# Gọi hàm để chạy chương trình
tinh_tong_phan_so_sn()
```
