**Bi 9. Tnh tng 2**

Tnh S(n) = 1^2 + 2^2 + 3^2 + ... + n^2 vi n l s nguyn khng m (0  n  10^5).

Cch lm: s dng cng thc tng bnh phng  bit:  S(n) = n(n+1)(2n+1) / 6.
- Thi gian: O(1) (mt vi php ton s hc).
- Ghi ch: vi Python khng lo trn s v int c  rng ty  ln.

V d: Input: 100000  Output: 333338333350000

```python

def sum_squares(n):
    return n * (n + 1) * (2 * n + 1) // 6

# V d kim th nhanh
print('sum_squares(0) =', sum_squares(0))
print('sum_squares(1) =', sum_squares(1))
print('sum_squares(10) =', sum_squares(10))  # 385
print('sum_squares(100000) =', sum_squares(100000))  # 333338333350000

# Nu dng trong bi thi/:
# n = int(input())
# print(sum_squares(n))
```
