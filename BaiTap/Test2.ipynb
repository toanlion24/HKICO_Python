```python
def tinh_tong_phan_so_Sn():
    while True: 
        input_str = input("Nhp s nguyn dng n: ")
        try:
            n = int(input_str)
            
            if n < 1: 
                print("Vui lng nhp s nguyn dng ln hn 0.")
            elif n > 10** 9:
                print("S qu ln, vui lng nhp s nguyn dng nh hn hoc bng 10^9.")
            else:
                break
        except ValueError:
            print("u vo khng hp l. Vui lng nhp mt s nguyn dng.")
            
    sn_value = n / (n+1)
    result_output = f"{sn_value:.2f}"
    print(f"Tng S_n = {result_output}")
    
tinh_tong_phan_so_Sn()
    
                
```

```python
try:
    n = int(input("Nhap so nguyen duong n: "))
except ValueError:
    print("Vui long nhap mot so nguyen duong.")
    exit(1)
    
if n < 1:
    print("Vui long nhap so nguyen duong lon hon 0.")
    exit(1)
if n%2 == 0:
    S_n = n // 2
else:
    S_n = -(n // 2) - 1
print(f"Gi tr ca n l: {n}")
print(f"Tng Sn l: {S_n}")
    
    
```
