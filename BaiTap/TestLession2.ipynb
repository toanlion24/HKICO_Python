```python
n = int(input("Nhp s n: "))
tong_le = 0
tong_chan = 0

# Lu li gi tr n ban u  in thng bo nu cn
n_bandau = n

while n > 0:
    chu_so = n % 10
    if chu_so % 2 != 0:
        tong_le += chu_so
    else:
        tong_chan += chu_so
    n //= 10

print(f"Tng cc ch s l ca {n_bandau} l: {tong_le}")
print(f"Tng cc ch s chn ca {n_bandau} l: {tong_chan}")
```
